```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            mean =37
            return mean
        elif Pclass==2:
            mean=29
            return mean
        else:
            mean = 24
            return mean
    else:
        return Age
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)
```
